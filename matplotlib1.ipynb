# **Histogram**

```python
                                                                                                                                                            marks=[10,20,30,40,50,60]
grade_intervals=[10,30,50,70]
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7)
plt.title("student_grade")
plt.xticks([10,30,50,70])
plt.xlabel("percentage")
plt.ylabel("no.of students")
plt.show()
```

```python
marks=[10,20,30,40,50,60]
grade_intervals=[10,30,50,70]
plt.hist(marks,grade_intervals,histtype="step",rwidth=0.7)
plt.title("student_grade")
plt.xticks([10,30,50,70])
plt.xlabel("percentage")
plt.ylabel("no.of students")
plt.show()
```

```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="b")
plt.title("student_grade")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no.of students")
plt.show()
```

```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7,facecolor="r")
plt.title("student_grade")
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("no.of students")
plt.show()
```

# **BAR PLOT**

```python
x=['R','A','V','I']
y = [30,50,70,90]
plt.bar(x,y , color = "orange" , width = 0.2) # by default it will take 0.8
plt.show(x,y)
```

```python
x=['R','A','V','I']
y = [30,50,70,90]
plt.barh(x,y , color = "pink" , height = 0.5) # by default it will take 0.8
plt.show(x,y)
```

```python
x = ["2020" , "2021" , "2022" , "2023"]
y = [280,350,470,590]
c = ["green","skyblue","pink","yellow"]
plt.bar(x,y,color=c,width=0.3)
plt.xlabel("Year")
plt.ylabel("no of students passed")
plt.title("Placemnts Data" , color="purple")
plt.show()
```

# **BOX PLOT**

> The shape of the boxplot shows how the dta is distributed and it also shows any outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list1)
plt.show()
```

```python

list2 = [1,2,2,2,3,5,6]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list2)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,5))
plt.boxplot(list1 , notch= True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list1 , vert= False)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , widths = 0.6)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["Abhi"] , widths = 0.5)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["Abhi"] , widths = 0.3 , patch_artist = True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["Abhi"] , widths = 0.3 , patch_artist = False, showmeans = True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["Abhi"] , widths = 0.3 , patch_artist = False, showmeans = True , boxprops = dict(color="r"))
plt.show()
```

# **PIECHART**

> A Pie chart is a circular statistical plot that can display only one series of data. The area of chart is the total percentage of given data



```python
import matplotlib.pyplot as plt
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values)
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 90)

plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance)
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 180)
# startangle means where it should represents
# counter plots
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.3,0,0,0])
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2])

plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2],shadow = True)

plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"])

plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"])

plt.legend(title = "Student_performance")
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"] , autopct = "%2.2f%%")

plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.figure(figsize=(9,10))
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"] , autopct = "%2.2f%%")
```
